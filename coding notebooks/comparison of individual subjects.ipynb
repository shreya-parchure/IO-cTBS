```python
#1 way ANOVA
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
ar = ['BME-01','BME-02','BME-03','BME-04','BME-05','BME-06','BME-07','BME-08','BME-10','BME-11','BME-13','BME-14','BME-15','BME-18','BME-20','BME-21','BME-22','BME-24','BME-25','BME-27','BME-28','BME-29']
num = np.size(ar)
for i in range(num):
  sub1 = df.loc[df['Subj_ID'] == ar[i]]
  sub1['lnMEP'] = pd.to_numeric(sub1['lnMEP'])
  formula = 'lnMEP ~ Time'
  model = ols(formula, sub1).fit()
  aov_table = sm.stats.anova_lm(model,type=1)
  print('one-way anova for raw median of lnMEP to time_point')
  print(ar[i])
  print(aov_table)




```

```python
#Unpaired t tests comparing baseline 30 MEPs to each post cTBS time point
import numpy as np
from scipy.stats import ttest_ind
from statsmodels.stats.multitest import multipletests


ar = ['BME-01','BME-02','BME-03','BME-04','BME-05','BME-06','BME-07','BME-08','BME-10','BME-11','BME-13','BME-14','BME-15','BME-18','BME-20','BME-21','BME-22','BME-24','BME-25','BME-27','BME-28','BME-29']
num = np.size(ar)
for i in range(num):
  sub1 = df.loc[df['Subj_ID'] == ar[i]]
  sub1['MEP'] = pd.to_numeric(sub1['MEP'])
  Base = sub1.loc[sub1['Time'] == 'Baseline']
  P0 = sub1.loc[sub1['Time'] == 'Post0']
  P10 = sub1.loc[sub1['Time'] == 'Post10']
  P20 = sub1.loc[sub1['Time'] == 'Post20']
  P30 = sub1.loc[sub1['Time'] == 'Post30']
  MEP_base = Base['MEP']
  MEP_P0 = P0['MEP']
  MEP_P10 = P10['MEP']
  MEP_P20 = P20['MEP']
  MEP_P30 = P30['MEP']
  alpha = 0.05
  P0_t_stat, P0_p_values = ttest_ind(MEP_base, MEP_P0)
  P0_reject, P0_p_values_corrected, _, _ = multipletests(P0_p_values, alpha=alpha, method='fdr_bh')

  P10_t_stat, P10_p_values = ttest_ind(MEP_base, MEP_P10)
  P10_reject, P10_p_values_corrected, _, _ = multipletests(P10_p_values, alpha=alpha, method='fdr_bh')

  P20_t_stat, P20_p_values = ttest_ind(MEP_base, MEP_P20)
  P20_reject, P20_p_values_corrected, _, _ = multipletests(P20_p_values, alpha=alpha, method='fdr_bh')

  P30_t_stat, P30_p_values = ttest_ind(MEP_base, MEP_P30)
  P30_reject, P30_p_values_corrected, _, _ = multipletests(P30_p_values, alpha=alpha, method='fdr_bh')

# Output results
  print(ar[i])
  print('T test of baseline to P0')
  print("Original p-values:", P0_p_values)
  print("Corrected p-values:", P0_p_values_corrected)
  print("Significant after FDR correction:", P0_reject)

  print('T test of baseline to P10')
  print("Original p-values:", P10_p_values)
  print("Corrected p-values:", P10_p_values_corrected)
  print("Significant after FDR correction:", P10_reject)

  print('T test of baseline to P20')
  print("Original p-values:", P20_p_values)
  print("Corrected p-values:", P20_p_values_corrected)
  print("Significant after FDR correction:", P20_reject)

  print('T test of baseline to P30')
  print("Original p-values:", P30_p_values)
  print("Corrected p-values:", P30_p_values_corrected)
  print("Significant after FDR correction:", P30_reject)

```

```python
#T test of MEPs vs all post cTBS MEPs
import numpy as np
from scipy.stats import ttest_ind
from statsmodels.stats.multitest import multipletests


ar = ['BME-01','BME-02','BME-03','BME-04','BME-05','BME-06','BME-07','BME-08','BME-10','BME-11','BME-13','BME-14','BME-15','BME-18','BME-20','BME-21','BME-22','BME-24','BME-25','BME-27','BME-28','BME-29']
num = np.size(ar)
for i in range(num):
  sub1 = df.loc[df['Subj_ID'] == ar[i]]
  sub1['MEP'] = pd.to_numeric(sub1['MEP'])
  Base = sub1.loc[sub1['Time'] == 'Baseline']
  P0 = sub1.loc[sub1['Time'] == 'Post0']
  P10 = sub1.loc[sub1['Time'] == 'Post10']
  P20 = sub1.loc[sub1['Time'] == 'Post20']
  P30 = sub1.loc[sub1['Time'] == 'Post30']
  MEP_base = Base['MEP']
  MEP_P0 = P0['MEP']
  MEP_P10 = P10['MEP']
  MEP_P20 = P20['MEP']
  MEP_P30 = P30['MEP']
  merged_array_column = pd.concat([MEP_P0, MEP_P10, MEP_P20, MEP_P30], axis=0).reset_index(drop=True)
  alpha = 0.05
  t_stat, p_values = ttest_ind(MEP_base, merged_array_column)
  reject, p_values_corrected, _, _ = multipletests(p_values, alpha=alpha, method='fdr_bh')


  print(ar[i])
  print('T test of baseline to all other time points')
  print("Original p-values:", p_values)
  print("Corrected p-values:", p_values_corrected)
  print("Significant after FDR correction:", reject)

```

```python
def sigmoid(x):
    return (1.8 / (1 + np.exp(-0.12*(x - 115)))) + 0.1
```

```python
#example plot MEP, MEP_max, MEP_130%
x_point = [80,90,100,110,115,120,125,130,140,150]
x_point = np.asarray(x_point)
x = np.linspace(80,150,1000)
y_point = sigmoid(x_point)
y = sigmoid(x)

plt.figure(figsize=(4,3.2),dpi=300)
plt.plot(x,y,color='black')
plt.scatter(x_point,y_point,color='black')
plt.xticks([70,100,130,160,190])

#plot arrow for S50
arrow_x = 115
arrow_y_start = sigmoid(arrow_x)
plt.annotate('', xy=(arrow_x, 0.1), xytext=(arrow_x, arrow_y_start), arrowprops=dict(arrowstyle='-|>', color='black', linestyle='-', lw=0.5))
plt.annotate('', xy=(70, arrow_y_start), xytext=(arrow_x, arrow_y_start), arrowprops=dict(arrowstyle='-', color='black', linestyle='--', lw=0.5))
plt.text(arrow_x, 0.05, '$S_{50}$', color='black', ha='center')

#plot line for MEPmax
arrow_x_max = 150
arrow_y_max = sigmoid(arrow_x_max)
arrow_x_end = 160
plt.annotate('', xy=(arrow_x_end, arrow_y_max), xytext=(70, arrow_y_max),
             arrowprops=dict(arrowstyle='-|>', color='black', linestyle='-', lw=0.5))
plt.text(160, arrow_y_max-0.005, '$MEP_{max}$', color='black', ha='left')

#add 130 RMT
arrow_x_130 = 130  # Starting x-coordinate for the arrow
arrow_y_max_130 = sigmoid(arrow_x_130)  # y-coordinate where the arrow will be placed
arrow_x_end_130 = 160  # Ending x-coordinate for the arrow
plt.annotate('', xy=(arrow_x_end_130, arrow_y_max_130), xytext=(70, arrow_y_max_130),
             arrowprops=dict(arrowstyle='-|>', color='black', linestyle='-', lw=0.5))
plt.annotate('', xy=(130, arrow_y_max_130), xytext=(130, 0),
             arrowprops=dict(arrowstyle='-', color='black', linestyle='--', lw=0.5))
plt.text(160, arrow_y_max_130-0.005, '$MEP_{130RMT}$', color='black', ha='left')


plt.xlabel('Intensity (%RMT)')
plt.ylabel('MEP amplitude (mV)')

plt.ylim(0, max(y)+0.15 )
plt.xlim(70,190)
plt.show()
```
